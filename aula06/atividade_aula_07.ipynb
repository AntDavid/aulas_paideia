```python
def cria_listas(dados_alunos):
    nomes = []
    idades = []
    matematica = []
    portugues = []
    ciencias = []
    status = []

    for nome, dados in dados_alunos.items():
        nomes.append(nome)
        idades.append(dados['idade'])
        matematica.append(dados['matematica'])
        portugues.append(dados['portugues'])
        ciencias.append(dados['ciencias'])
        status.append(dados['status'])
    
    print('Nomes:', nomes)
    print('Idades:', idades)
    print('Notas de matemática:', matematica)
    print('Notas de português:', portugues)
    print('Notas de ciências:', ciencias)
    print('Status:', status)
    print('\n')
    return matematica, portugues, ciencias


def verifica_status(dados_alunos):
    aprovados = []
    reprovados = []

    for nome, dados in dados_alunos.items():
        if dados['status']=='Aprovado' or dados['status']=='Aprovada':
            aprovados.append(nome)
        else:
            reprovados.append(nome)

    print('Aprovados:', aprovados)
    print('Reprovados:', reprovados)
    print('\n')

def calcula_score_das_disciplinas (portugues, matematica, ciencias):
    score_portugues = sum(portugues)/len(portugues)
    score_matematica = sum(matematica)/len(matematica)
    score_ciencias = sum(ciencias)/len(ciencias)

    print('Score de ciências:', score_ciencias)
    print('Score de português:', score_portugues)
    print('Score de matemática:', score_matematica)
    print('\n')


def verifica_a_maior_nota(dados_alunos):
    maior_matematica = max(dados_alunos, key=lambda nome: dados_alunos[nome]['matematica'])
    maior_portugues = max(dados_alunos, key=lambda nome: dados_alunos[nome]['portugues'])
    maior_ciencias = max(dados_alunos, key=lambda nome: dados_alunos[nome]['ciencias'])
    
    print('Maior nota em matemática:', maior_matematica)
    print('Maior nota em português:', maior_portugues)
    print('Maior nota em ciências:', maior_ciencias)


if __name__ == '__main__':
    dados_alunos = {
    'Aline': {'idade': 12, 'matematica': 8.5, 'portugues': 7.0, 'ciencias': 9.0, 'status': 'Aprovada'},
    'Antonio': {'idade': 13, 'matematica': 6.0, 'portugues': 8.0, 'ciencias': 7.5, 'status': 'Reprovado'},
    'Yan': {'idade': 12, 'matematica': 9.0, 'portugues': 9.5, 'ciencias': 10.0, 'status': 'Aprovada'},
    'Daniel': {'idade': 14, 'matematica': 5.5, 'portugues': 6.0, 'ciencias': 6.5, 'status': 'Reprovado'},
    'Pedro': {'idade': 13, 'matematica': 7.0, 'portugues': 8.5, 'ciencias': 7.5, 'status': 'Aprovada'},
    'Vitor': {'idade': 12, 'matematica': 10.0, 'portugues': 7.5, 'ciencias': 8.0, 'status': 'Aprovada'},
    'Vinícius': {'idade': 14, 'matematica': 4.5, 'portugues': 6.0, 'ciencias': 5.0, 'status': 'Reprovado'},
    'Keezya': {'idade': 13, 'matematica': 8.0, 'portugues': 9.0, 'ciencias': 7.0, 'status': 'Aprovado'},
    'Isabela': {'idade': 12, 'matematica': 9.5, 'portugues': 8.0, 'ciencias': 9.0, 'status': 'Aprovada'},
    'João': {'idade': 13, 'matematica': 5.0, 'portugues': 7.0, 'ciencias': 6.0, 'status': 'Reprovado'}
}
    
    portugues, matematica, ciencias = cria_listas(dados_alunos)
    cria_listas(dados_alunos)
    verifica_status(dados_alunos)
    calcula_score_das_disciplinas(portugues, matematica, ciencias)
    verifica_a_maior_nota(dados_alunos)
```
